Observe que o mapeamento da esta√ß√£o est√° configurado como uma fun√ß√£o, `create_station_mapping()`. Uma vez que √© poss√≠vel que mais esta√ß√µes sejam adicionadas ou removidas ao longo do tempo, esta fun√ß√£o nos permitir√° combinar as informa√ß√µes da esta√ß√£o em todas as tr√™s partes dos nossos dados quando estivermos prontos para explorar tudo.

```python
# 4.1.2 - Mostra as primeiras linhas do arquivo de dados das esta√ß√µes
station_info = pd.read_csv('201402_station_data.csv')
station_info.head()
```

4.1.3 - Preencha a fun√ß√£o abaixo de forma que a fun√ß√£o retorne um mapeamento entre o id da esta√ß√£o (`station_id`) e a cidade em que ela se encontra (`landmark`).

```python
def create_station_mapping(station_data):
    # Inicializa a vari√°vel station_map como um dicion√°rio vazio.
    station_map = dict()
    for data_file in station_data:
        # Prepara para a leitura do arquivo de dados 'station_data'.
        with open(data_file, 'r') as f_in:
            weather_reader = csv.DictReader(f_in)
            # Inicia mapeamento entre o Id da esta√ß√£o e sua cidade correspondente.
            for row in weather_reader:
                st_id = row['station_id']
                land = row['landmark']
                station_map[st_id] = land
    return station_map
```

4.1.4 - Voc√™ pode agora usar o mapeamento para condensar as viagens para as colunas selecionadas acima. Isto acontecer√° na fun√ß√£o abaixo `summarise_data()`. Nela o m√≥dulo `datetime` √© usado para fazer o **p**arse do tempo (timestamp) em formato de strings no arquivo original para um objeto usando a fun√ß√£o `strptime`. Este objeto permitir√° a convers√£o para outros **f**ormatos de datas usando a fun√ß√£o `strftime`. O objeto possui tamb√©m outras fun√ß√µes que facilitam a manipula√ß√£o dos dados. Veja [este tutorial](http://usandopython.com.br/manipulando-data-hora-python-datetime/) para entender um pouco melhor como trabalhar com a biblioteca.

Voc√™ precisa concluir duas tarefas para completar a fun√ß√£o `summarise_data()`. Inicialmente, voc√™ dever√° realizar a opera√ß√£o de converter a dura√ß√£o das viagens de segundos para minutos. Esta √© muito f√°cil, pois existem 60 segundos em um minuto!

Na sequ√™ncia, voc√™ deve criar colunas para o ano, m√™s, hora e dia da semana. Verifique o tutorial acima ou a [documenta√ß√£o para o objeto de datetime no m√≥dulo datetime](https://docs.python.org/2/library/datetime.html#datetime-objects). 

** TODO: Encontre os atributos e m√©todos necess√°rios para poder completar o c√≥digo abaixo **

*Dica*: Voc√™ pode abrir uma nova caixa para testar um peda√ßo do c√≥digo ou verificar uma vari√°vel que seja global. Caso ela esteja dentro da fun√ß√£o, voc√™ tamb√©m pode usar o comando `print()` para imprimi-la e ajudar no Debug.

```python
def summarise_data(trip_in, station_data, trip_out):
    # gera o dicion√°rio de mapeamento entre esta√ß√µes e cidades
    station_map = create_station_mapping(station_data)
    with open(trip_out, 'w') as f_out:
        # configura o objeto de escrita de csv       
        out_colnames = ['duration', 'start_date', 'start_year',
                        'start_month', 'start_hour', 'weekday',
                        'start_city', 'end_city', 'subscription_type']        
        trip_writer = csv.DictWriter(f_out, fieldnames = out_colnames)
        trip_writer.writeheader()
        for data_file in trip_in:
            with open(data_file, 'r') as f_in:
                # configura o leitor do csv
                trip_reader = csv.DictReader(f_in)
                # processa cada linha lendo uma a uma
                for row in trip_reader:
                    # Inicia vari√°vel new_point como um dicion√°rio vazio.
                    new_point = {}
                    
                    # converte a dura√ß√£o de segundos para minutos.
                    ### TODO: Pergunta 3a: Adicione uma opera√ß√£o matem√°tica ###
                    ### para converter a dura√ß√£o de segundos para minutos.  ###
                    new_point['duration'] = float(row['Duration'])/60
                    
                    # refer√™ncia usada: semana 3, Aula 1, item 7, 'Arrumando os tipos de dados'
                    # https://goo.gl/tHMCS7
                    
                    # reformate strings com datas para m√∫ltiplas colunas
                    ### TODO: Pergunta 3b: Preencha os __ abaixo para criar os  ###
                    ### campos experados nas colunas (olhe pelo nome da coluna) ###
                    trip_date = datetime.strptime(row['Start Date'], '%m/%d/%Y %H:%M')
                    new_point['start_date']  = trip_date.day
                    new_point['start_year']  = trip_date.year
                    new_point['start_month'] = trip_date.month
                    new_point['start_hour']  = trip_date.hour
                    new_point['weekday']     = trip_date.weekday()
                    # refer√™ncia usada: https://docs.python.org/2/library/datetime.html
                    
                    # TODO: mapeia o terminal de inicio e fim com o a cidade de inicio e fim
                    new_point['start_city']  = station_map[row['Start Terminal']]
                    new_point['end_city']    = station_map[row['End Terminal']]
                    
                    # refer√™ncia: semana 3, Aula 1, item 7,'Arrumando os tipos de dados'
                    # notebook: l1-starter-code.ipynb | https://goo.gl/GxJ18M
                    
                    # TODO: existem dois nomes diferentes para o mesmo campo. Trate cada um deles.
                    if 'Subscription Type' in row:
                        new_point['subscription_type'] = row['Subscription Type']
                    else:
                        new_point['subscription_type'] = row['Subscriber Type']
                    # refer√™ncia: semana 3, Aula 1, item 7,'Arrumando os tipos de dados'
                    # notebook: l1-starter-code.ipynb | https://goo.gl/GxJ18M
                    
                    # escreve a informa√ß√£o processada para o arquivo de sa√≠da.
                    trip_writer.writerow(new_point)
```
